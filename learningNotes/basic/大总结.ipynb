# 变量和简单的数据类型
- 变量名只能包含字母、数字和下划线，只能使用数字和下划线开头。
- 不能包含空格
- 不要和关键字和函数名重名
- 变量名应简短且具有描述性
- 慎用小写字母l和大写字母O
- `\n`换行符 `\t`制表符
- `strip()`删除左右两端空白，`rstrip()`删除右端空白,`lstrip()`删除左端空格
- 全部大写：`name.title()`全小写：`name.lower()` 首字母大写：`name.title()`
- 在书写很大的数时，可使用_来将其分组，使其更好阅读
- 可在一行给多变量进行赋值`(x,y,z = 0,0,0)`
- 常量：一般使用全大写字母，单词使用_ `(MAX_CONSTANT)`
- 数：
    - 计算符号：加(+)、减(-)、乘(*)、除(/)、乘方(**),多种运算可以使用()来调整运算的顺序
    - 任意两个数相除结果都是浮点数，即便是两个整数且能整除
    - 在任何的运算中，只要有操作数是浮点数，默认结果是浮点数
    - 对于浮点数保留小数的操作：
        ```python
        a = 8.9+3.2485
        print("%.2f" % a)
        print(round(a, 2))
        #下面两种为转换成str的操作
        print("{:.2f}".format(a))  
        print(f"{a:.2f}") 
        ```
    - 求模运算符(%)：指出余数是多少，如果一个数可被另一个数整除，余数就为0，放回的结果就为0




# 列表
- 查看：访问列表元素：`list[索引]`  (都是从0开始，-1是最后一个)
- 增加：
    - 插入 `insert(索引，'元素')`
    - 末尾追加 `append('元素')`
- 修改：修改 `列表[] = "元素"`
- 删除：
    - 当需要记录当前被删除的元素时使用`pop(索引)`,不添加索引就是最后一个,不需要记录可以直接使用`del list[索引]`
    - 根据值删除元素，使用`remove("值")`方法,需要注意，`remove()`只删除第一个指定的值，若需删除多个，则需要使用循环进行删除
    - 清空列表：`list.clear()`或`del list`
- 排序：
    - 永久排序`sort()`
    - 临时排序`sorted()`
    - 按字母的相反顺序排序：`sort(reverse = True)`
- list.reverse：当前排序的逆序（即反向输出），注意：reverse也是永久的
- 列表长度，`len()`  

- 遍历整个列表可以使用for循环
- 创建数值列表：使用`range()`函数,如：`list(range(1,6))`
- 数值列表计算：最大值`max()`，最小值`min()`，总和`sum()`, 如果需要其他的数值信息可以添加numpy库进行操作(如：np.mean())
- 列表推导式：`list = [value**2 for value in range(1,11)]`

- 列表切片：`list[索引]  `
    - 列表的全部:`list[:]`
    - 列表的第一个和第二元素：`list[1:3]`
    - 列表的前四个元素：`list[:4]`
    - 列表的最后三个元素；`list[-3:]`
- 复制列表:  
    - 复制列表还可以分别对其进行操作：`list1 = list2[:]`
    - 全部复制，后续对于两个列表不同操作都会同时进行，两个列表始终都保持一致：`list1 = list2`

- 如果需要储存一组在程序的整个生命周期内都不会改变的值，就可以使用元组。`dimensions = (200,90)`,如果需要修改即重新定义

# if语句
- 条件测试：结果显示为True或False  
    - 检查是否相等：`==`
    - 检查是否不等：`!=`
    - 数值比较：`==,!=,>,>=,<,<=`
    - 检查多个条件:   
        - 两个同时满足：`and ` 
        - 一个满足就行：`or  `
    - 检查特等值是否在列表中：`in  `
    - 检查特等值是否不在列表中：`not in`
- if语句：
    - if语句:
    ```python
        if 条件:
            满足条件的操作(如print("满足条件"))
    ```
    - if-else语句：
    ```python
        if 条件:
            满足条件的操作(如print("满足条件"))
        else:
            不满足条件的操作(如print("满足条件"))
    ```
    - if-elif-else语句：(检查两个条件以上的情形)<br>
    一个例子：游乐场买票：4岁以下免费。4（含）-18岁收费$25。18岁以上收费$40
    ```python
        if age < 4:
            price = 0
        elif age < 18:
            price = 25
        else:
            price = 40
        print(f"Your admission cost is {price}.")
    ```
    **注意：可以使用多个elif代码块，也可以省略else代码块，但仅适用于只有一个条件满足的情况**<br>
    **总之，如果只想运行一个代码块，就是用if-elif-else语句；如果要运行多个代码块，就是用一些列独立的if语句**

- if语句处理列表
    - 能够检查列表中元素：`if list = '特殊值'`
    - 确定列表非空：`if list: `<br>当list为空时返回的False，添加输出条件即可
    - 使用多个列表

# 字典
- 创建字典：dict={'key1':value1,'key2':value2,....} <br>**注意：集合也可以用{}来进行创建，区别：集合没有键值对应，且不会以特定的顺序来储存元素。**
- 查看：
    - 访问具体值：`dict['key']`
    - 查看全部的键：`dict.keys()`
    - 查看全部的值：`dict.values()`
    - 查看全部内容(键对值)：`dict.items()`
    - 遍历所有的键对值：
        ```python
        for key,value in dict.items():
            print(f"\nKey:{key}")
            print(f"\nValue:{value}")
        ```
    - 遍历所有的键：
        ```python
        for key in dict.keys():
            print(f"\nKey:{key}")
        ```
        - 当需要添加特定顺序时可以添加sorted：`for key in sorted(dict.keys()):`
        - 当需要判断指定的键是否在其中，或对特定的键输出特殊的语句，可以添加if语句进行判断
    - 遍历所有的值：
        ```python
        for value in dict.values():
            print(f"\nValue:{value}")
        ```
        - 为了剔除重复项，可以使用集合(set)：`for value in set(dict.values()):`
    - 使用get()访问值：当字典中没有要访问的键值时会引发报错，可以使用get()方法来处理
        - `get('指定不存在的键','不存在时返回的值')`
        - `dict.get('point','No point value assigned')`
        - 当get()的第二个参数即没有设置返回值时，会直接返回None

- 添加：添加键对值：`dict['key'] = value`
- 修改：修改字典中的值：`dict['key'] = value_change`
- 删除：删除键值对：使用del语句将相应的键值对**彻底删除**，**在使用del语句时，必须指定字典名和要删除的键**  
        `del dict['key']`
- 嵌套：
    - 字母列表：
        ```python
        dict1 = {'key1':value1,'key2':value2,....}
        dict2 = {'key1':value1,'key2':value2,....}
        dict3 = {'key1':value1,'key2':value2,....}
        dicts = [dict1,dict2,dict3]
        for dict in dicts:
            print(dict['key1'],dict['key2'],...)
        ```
    - 在字典中存储列表： *对于字典中的列表需要一个for循环来输出==>需要两个for循环展示全部信息*
        ```python
        dict1 = {'key1':value1,'key2':[value2,value3,value4]....}
        for key,values in dict1['key2']:
            print(f"{key}")
            for value in values:
                print(f"{value}")
        ```
    - 在字典中存储字典：

# while循环
- while循环：不断地运行，直到指定的条件不在满足为止
    ```python
    while "满足的条件":
        满足条件的操作
    ```
- 使用标志来运行程序，当flag=True是进行循环，当False时结束循环
    ```python
    while flag:
        message = input(prompt)
        if message == 'quit':
            flag = False
        else:
            print(message)
    ```
- 使用break退出循环：**立即跳出循环**，不在执行后面的代码
    ```python
    while True:
        city = input(prompt)
        if city == 'quit':
            break
        else:
            print(city)
    ```
    **注意：在所以的python循环中都可以使用break语句**

- 在循环中使用continue：**返回循环的开头**，只跳出当前循环轮次
    ```python
    num = 1
    while num < 10:
        num += 1
        if num % 2 ==0:
            continue
        print(num)
    ```
- 如果陷入无限循环可以按Ctrl+C
- 使用while循环处理列表和字典
    - 在列表之间移动元素：一个有元素的列表，一个空列表，每循环一次将有元素的列表元素移至空列表，并在原有列表中删除，循环结束的条件是原有列表为空。*（练习7.8）*
    - 删除为特定值的所有列表元素：使用remove()方法来删除列表中的特定值，如删除一个列表中的多个重复值 *（练习7.9）*
    - 使用用户输入填充字典：建立一个空的字典，并设置一个标志用于指出调查是否继续，在循环需要输入两个值，一个作为字典的键，一个作为值进行储存。*（练习7.10）*


# 函数
- 定义函数：
    ```python
    def function(username):
        函数体
    ```
- 实参和形参：
    - 实参，就是调用函数括号当中的参数，`fuction('Mary')`，Mary就是实参
    - 形参，就是定义函数括号当中的参数，`def fuction(username)`，其中username就是形参
- 传递实参
    - 位置实参：实参与形参的顺序相同
    - 关键字实参：每个实参都由变量名和值组成，**确保定义函数时有相同的形参**
    - 默认值，可以在定义函数时给没给形参指定默认值，当使用默认值时，**必须在形参列表中先列出没有默认值的形参，再列出有默认值的形参
        ```python
        def describe_city(city,country='chile'):
            msg = f"{city.title()} is in {country.title()}"
            print(msg)
        ```
- 返回值：return

    - 函数的返回值可以是一个或一组值，也可以是一个字典,列表等复杂对象
    - 每个函数都应只负责一项具体的工作，同时可以在一个函数中调用一个函数
- 传递任意数量的实参：
    - *toppings,可以收入任意数量的实参，其实是创建了一个叫toppings的元组
        ```python
        def make_pizza(*toppings):
            print(toppings)
        make_pizza('pepperoni')
        make_pizza('mushrooms','green peppers')
        ```
    - **user_info,能够接受任意数量的键值对（关键字实参），
        ```python

            user_info['first_name'] = first
            user_info['last_name'] = last
            return user_info
        user_profile = build_profile('albert','einsten',location='princeton',field='physics')
        print(user_profile)
        ```
- 将函数以独立模块的形式进行储存的调用方式，使用import将已写好的模块(.py文件)导入,在需要使用函数时直接调用需要的函数
    - 导入整个模块：`import module_name`
    - 导入特定的函数：`from module_name import fuction_0,fuction_1,fuction_2`**注意fuction_name没有()**
    - 使用as给函数指定别名：`from module_name import fuction_0 as fn_0`
    - 使用as给模块指定别名：`import module_name as md`
    - 导入模块中的所有函数：`from module_name improt *`
- 函数编写的一些注意事项：
    - 应给函数指定描述性名称，且只使用小写字母和下划线
    - 在给形参指定默认值，或者调用函数函数中的关键字实参时，等号两边不应该有空格
    - 如果程序或模块包含多个函数，可以使用两个空行将相邻的函数分开
