```python
import pandas as pd
```

```python
# 读取文件，并确保股票代码列是字符串类型
df_1 = pd.read_excel('发展能力.xlsx', dtype={'股票代码': str})
df_2 = pd.read_excel('管理层治理能力.xlsx', dtype={'股票代码': str})
df_3 = pd.read_excel('研发投入情况表.xlsx', dtype={'股票代码': str})
df_4 = pd.read_excel('主要财务指标.xlsx', dtype={'股票代码': str})
df_5 = pd.read_excel('资产负债表.xlsx', dtype={'股票代码': str})
df_6 = pd.read_excel('公司基本信息表.xlsx', dtype={'股票代码': str})
```

```python
df_1
```

```python
# 合并前五个文件，假设它们都有'股票代码'和'截止日期'两列
df_combined = pd.merge(df_1, df_2, on=['股票代码', '截止日期'])
df_combined = pd.merge(df_combined, df_3, on=['股票代码', '截止日期'])
df_combined = pd.merge(df_combined, df_4, on=['股票代码', '截止日期'])
df_combined = pd.merge(df_combined, df_5, on=['股票代码', '截止日期'])#
df_combined
```

```python
df_final = pd.merge(df_combined, df_6, on='股票代码', how='left')  # 使用left join以保留所有df_combined中的行
df_final
```

```python
# 保存合并后的文件
df_final.to_excel('合并后的文件.xlsx', index=False)
```
