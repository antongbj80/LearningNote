```python
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
# 读取文件，并确保股票代码列是字符串类型
df = pd.read_excel('附件1操作数据.xlsx')
df
```

```python
# 统计不同籍贯的个数
Provence_counts = df['籍贯'].value_counts().reset_index().rename(columns={'index': '籍贯', '籍贯': '计数'})
pro_df = pd.merge(Provence_counts, df[['籍贯','城市', '经度', '纬度']], on='籍贯')
pro_df.drop_duplicates(inplace=True)
pro_df
```

```python
# 统计不同城市的个数
city_counts = df['城市'].value_counts().reset_index().rename(columns={'index': '城市', '城市': '计数'})
city_df = pd.merge(city_counts, df[['城市', '经度', '纬度']], on='城市')
city_df.drop_duplicates(inplace=True)
city_df
```

```python
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType
import pandas as pd  # 如果你使用Pandas处理数据

geo = Geo()
geo.add_schema(maptype='china', itemstyle_opts=opts.ItemStyleOpts(color='white', border_color='black'),
               label_opts=opts.LabelOpts(is_show=True))
#设置地图类型，背景和边界颜色
geo.add("geo", pro_df[['籍贯','计数']].values.tolist(), type_=ChartType.EFFECT_SCATTER)
#添加图表类型，传入数据等
geo.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
#系列配置项，忽略标签
geo.set_global_opts(
    visualmap_opts=opts.VisualMapOpts(is_piecewise=True,
    pieces=[{"min":0, "max":30, "label":"0-30人", "color":"pink"},
            {"min":31, "max":60, "label":"31-60人", "color":"cyan"},
            {"min":61, "max":90, "label":"61-90人", "color":"yellow"},
            {"min":91, "max":120, "label":"91-120人", "color":"blue"},
            {"min":121, "max":150, "label":"121-150人", "color":"red"}
            ]
            ),
            title_opts=opts.TitleOpts(title="救援人员籍贯分布"),
            )
#全局配置项，设置动画、颜色、标题等
geo.render("各省人口数1.html")
#把制作完成的图表输出为html文件
# 使用webbrowser库打开HTML文件  
geo.render_notebook()
```

```python
from pyecharts.charts import Map
from pyecharts import options as opts

# 假设你已经有了数据，这里是模拟数据
s = df['籍贯'].value_counts()
data_pair = [[province, int(s[province])] for province in s.index]
print(data_pair)


# 创建地图对象
map = Map(init_opts=opts.InitOpts(bg_color='white', width='1200px', height='700px'))
# 添加地图数据
map.add("", data_pair, "china")

# 设置全局配置项
map.set_global_opts(
    title_opts=opts.TitleOpts(title='救援人员人口分布', pos_left='200', pos_top='10',
                                  title_textstyle_opts=opts.TextStyleOpts(color='black', font_size=16)),
        visualmap_opts=opts.VisualMapOpts(type_='color',range_color=['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8'],
                                          max_=150, is_piecewise=True,pos_left='100', pos_bottom='100',
                                          textstyle_opts=opts.TextStyleOpts(color='black', font_size=16))
)

# 渲染图表到 HTML 文件中，也可以直接使用 render_notebook() 在 Jupyter Notebook 中显示
map.render('rescue_personnel_distribution.html')
```
